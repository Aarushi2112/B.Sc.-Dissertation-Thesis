```python
import pandas as pb
x=pb.read_csv('Period_rotator_jyi.csv')
x.head()
```

```python
import seaborn as sns
import numpy as np
import matplotlib.ticker as plticker

import matplotlib.pyplot as plt
plt.figure(figsize=(15,7))
plt.rcParams.update({'font.size':22})
plt.rcParams.update({'font.sans-serif':'Helvetica'})
print(plt.rcParams['font.family'])
print(plt.rcParams['font.sans-serif'])
plt.suptitle('Rotators (n=45)',fontsize=30)



plt.subplot(1,2,1)
sns.histplot(x='Period',data=x,bins=10)
plt.ylabel("Count", size=28)
plt.xlabel("Period (<1 day)", size=28)
ax=plt.gca()
ax.margins(x=0.01)
loc = plticker.MultipleLocator(base=0.1) # this locator puts ticks at regular intervals
ax.xaxis.set_major_locator(loc)
ax.set_ylim([0,15])
plt.xlim(right=1)


xs = np.linspace(0, 0.1, 1)
plt.hlines(y=[2,1], xmin=[0,0], xmax=[len(xs)], colors='red', linestyles='--', lw=2)
ax.annotate("", xy=(0.15, 2), xytext=(0.2, 12), arrowprops=dict(arrowstyle="->"))
plt.text(0.2, 12, 'J114119.13-344444.0 (p=0.118),\nJ092654.45-361238.1 (p=0.135)', fontsize = 13)
ax.annotate("", xy=(0.2, 2), xytext=(0.25, 10), arrowprops=dict(arrowstyle="->"))
plt.text(0.25,10, 'J133722.08+413437.7 (p=0.221), \nJ122721.71-341913.7 (p=0.274)', fontsize = 13)
ax.annotate("", xy=(0.26, 1), xytext=(0.31, 8), arrowprops=dict(arrowstyle="->"))
plt.text(0.31, 8, 'J122721.71-341913.7 (p=0.274)', fontsize = 13)
ax.annotate("", xy=(0.38, 2), xytext=(0.42, 6), arrowprops=dict(arrowstyle="->"))
plt.text(0.42, 6, 'J104354.15-290302.7 (p=0.363), \nJ083348.25+574833.7 (p=0.371)', fontsize = 13)
extraticks=[1]
ax.set_yticks(list(ax.get_yticks()) + extraticks)

plt.rcParams.update({'font.size':22})
plt.subplot(1,2,2)
sns.histplot(x='Periods',data=x,bins=10)
plt.ylabel("Count", size=28)
plt.xlabel("Period (>1 day)", size=28)
ax=plt.gca()
loc = plticker.MultipleLocator(base=10) # this locator puts ticks at regular intervals
ax.xaxis.set_major_locator(loc)
ax.set_ylim([0,16])
ax.set_xlim([1,70])
x_ticks = ax.xaxis.get_major_ticks()
x_ticks[0].label1.set_visible(False)
ax.margins(x=0)

plt.savefig('my_plot.png')
```
