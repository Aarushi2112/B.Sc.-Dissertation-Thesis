```python
import pandas as pb

x=pb.read_csv('radec.csv')
x.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
fig=plt.gcf()
fig.set_size_inches(12,8)
plt.rcParams.update({'font.size':20})
plt.rcParams.update({'font.sans-serif':'Helvetica'})
print(plt.rcParams['font.family'])
print(plt.rcParams['font.sans-serif'])

sns.color_palette("Blues", as_cmap=True)
sns.scatterplot(x='Right Ascension (deg)',y='Declination (deg)',data=x,hue='Classification',palette='CMRmap_r',size='Magnitude (V)',sizes=(0.1,100));
plt.ylim(-80,80)
plt.xlim(250,50)
plt.grid()

plt.xlabel("Right Ascension (deg)", size=25)
plt.ylabel("Declination (deg)", size=25)


plt.title("Equatorial Coordinates of Classified Variable Stars", size=27)
plt.show()

```
