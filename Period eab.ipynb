```python
import pandas as pb
x=pb.read_csv('Period_eab_jyi.csv')
x.head()
```

```python
import seaborn as sns
import numpy as np
import matplotlib.ticker as plticker

import matplotlib.pyplot as plt
plt.figure(figsize=(7,7))
plt.rcParams.update({'font.size':22})
plt.rcParams.update({'font.sans-serif':'Helvetica'})
print(plt.rcParams['font.family'])
print(plt.rcParams['font.sans-serif'])
plt.suptitle('EA & EB Eclipsing Binaries (n=60)',fontsize=29)




sns.histplot(x='Period',data=x,bins=10)
plt.ylabel("Count", size=28)
plt.xlabel("Period (days)", size=28)
ax=plt.gca()
ax.margins(x=0.01)
loc = plticker.MultipleLocator(base=1) # this locator puts ticks at regular intervals
ax.xaxis.set_major_locator(loc)
ax.set_xlim([0,11])
#plt.xlim(right=11)


xs = np.linspace(0, 1, 11)
plt.hlines(y=[1], xmin=[0], xmax=[len(xs)], colors='red', linestyles='--', lw=2)
ax.annotate("", xy=(10.2, 1), xytext=(6, 20), arrowprops=dict(arrowstyle="->"))
plt.text(4.3, 21, 'J143549.75-180224.1 (p=10.72)', fontsize = 13)

extraticks=[1]
ax.set_yticks(list(ax.get_yticks()) + extraticks)
y_ticks = ax.yaxis.get_major_ticks()
y_ticks[0].label1.set_visible(False)


plt.savefig('my_plot.png')
```
