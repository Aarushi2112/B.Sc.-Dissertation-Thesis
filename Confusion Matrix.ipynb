```python
import pandas as pb
x=pb.read_csv('confusion matrix.csv')
x.head()
```

```python
data=x.pivot(index='Classification from SVS Lightcurves',columns='Classification under various Astronomical Surveys',values='Percentage')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.rcParams.update({'font.sans-serif':'Helvetica'})
print(plt.rcParams['font.family'])
print(plt.rcParams['font.sans-serif'])
fig=plt.gcf()
fig.set_size_inches(10,7)
sns.set(font_scale=2.7)
ax=plt.axes()
cax = plt.gcf().axes[-1]
cax.tick_params(labelsize=20)
display_data=sns.heatmap(data,xticklabels=True,yticklabels=True, annot=True,annot_kws={"size": 25}, fmt=".2f",linewidths=.5,cmap="Blues",ax=ax,vmin=0,vmax=100)
display_data.set_xticklabels(display_data.get_xticklabels(), rotation=30)
display_data.set_yticklabels(display_data.get_yticklabels(), rotation=0)
ax.figure.axes[-1].set_ylabel('% of Match', size=25)
#ax.set_title('Confusion Matrix',size=30)
plt.ylabel("Classification from SVS Lightcurves", size=25)
plt.xlabel("Classification from Vizier Catalogue Service", size=25)
plt.show()

figure = display_data.get_figure()
#figure.savefig('Confusion Matrix.jpg', dpi=300)
```
