```python
import pandas as pb

x=pb.read_csv('photometry_all.csv')
x.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
fig=plt.gcf()
fig.set_size_inches(12,8)

sns.color_palette("Blues", as_cmap=True)
sns.scatterplot(x='g-r',y='r-i',data=x,hue='Classification',s=30,style='Classification');
plt.grid()
plt.xlabel("g-r", size=20)
plt.ylabel("r-i", size=20)


plt.show()
```
